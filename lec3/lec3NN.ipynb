```python
import numpy as np

class NeuralNet:
    def __init__(self, input_size, hidden_size, output_size):
        np.random.seed(42)
        self.W1 = np.random.uniform(size=(input_size, hidden_size))
        self.b1 = np.random.uniform(size=(1, hidden_size))
        self.W2 = np.random.uniform(size=(hidden_size, output_size))
        self.b2 = np.random.uniform(size=(1, output_size))

        self.predicted_output = None
        self.hidden_output = None

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def feedforward(self, x):
        self.hidden_input = np.dot(x, self.W1) + self.b1
        self.hidden_output = self.sigmoid(self.hidden_input)

        self.final_input = np.dot(self.hidden_output, self.W2) + self.b2
        self.predicted_output = self.sigmoid(self.final_input)
        return self.predicted_output

    def backprop(self, x, y, learning_rate):
        error = y - self.predicted_output
        d_predicted_output = error * self.sigmoid_derivative(self.predicted_output)

        error_hidden_layer = d_predicted_output.dot(self.W2.T)
        d_hidden_layer = error_hidden_layer * self.sigmoid_derivative(self.hidden_output)

        self.W2 += self.hidden_output.T.dot(d_predicted_output) * learning_rate
        self.b2 += np.sum(d_predicted_output, axis=0, keepdims=True) * learning_rate
        self.W1 += x.T.dot(d_hidden_layer) * learning_rate
        self.b1 += np.sum(d_hidden_layer, axis=0, keepdims=True) * learning_rate

    def train(self, x, y, epochs=10000, lr=0.1):
        for epoch in range(epochs):
            self.feedforward(x)
            self.backprop(x, y, lr)

            if (epoch + 1) % 1000 == 0:
                loss = np.mean(np.square(y - self.predicted_output))
                print(f'epoch {epoch+1}, loss: {loss.item():.4f}')

if __name__ == "__main__":
    x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    y = np.array([[0], [1], [1], [0]])

    nn = NeuralNet(2, 2, 1)
    nn.train(x, y, epochs=10000, lr=0.1)

    print("\nFinal Prediction:")
    print(nn.feedforward(x))
```
